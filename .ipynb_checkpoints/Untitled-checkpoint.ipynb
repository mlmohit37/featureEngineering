```python
# You can use inbuilt functions like groupby and aggregations like sum(), cumsum() to merge some rows based on their 1. Trip_uuid, Source ID and Destination ID 2. Further aggregate on the basis of just Trip_uuid. 
# You can also keep the first and last values for some numeric/categorical fields if aggregating them wonâ€™t make sense.

dic = {'source_name':'first', 'osrm_distance':'sum', 'trip_creation_time':'first', 'route_type':'first',
       'destination_name':'first', 'od_start_time':'first', 'od_end_time':'first',
       'start_scan_to_end_scan':'sum', 'actual_distance_to_destination':'sum', 'actual_time':'sum',
       'osrm_time':'sum', 'segment_actual_time':'sum',
       'segment_osrm_time':'sum','segment_osrm_distance':'sum'}

df = df.groupby(['trip_uuid', 'source_center', 'destination_center'], as_index=False).agg(dic) 
df
```

## 2> Build some features to prepare the data for actual analysis. Extract features from the below fields:
    a. Destination Name: Split and extract features out of destination. City-place-code (State)
    b. Source Name: Split and extract features out of destination. City-place-code (State)
    c. Trip_creation_time: Extract features like month, year and day etc

#### 2.a. Destination Name: Split and extract features out of destination. City-place-code (State)
#### 2.b. Source Name: Split and extract features out of destination. City-place-code (State)

```python
def city(a):
    m = a.split("_")
    try:
        return m[0]
    except:
        return ' '

def place(b):
    n = b.split("_")
    try:
        return n[1]
    except:
        return ' '
def code(c):
    o = c.split("_")
    try:
        return o[2]
    except:
        return o[-1]
```

```python
df['des_city'] = df['destination_name'].apply(city)
df['des_place'] = df['destination_name'].apply(place)
df['des_code'] = df['destination_name'].apply(code)
df['sor_city'] = df['source_name'].apply(city)
df['sor_place'] = df['source_name'].apply(place)
df['sor_code'] = df['source_name'].apply(code)
```

```python
df.head()
```

#### 2.g. Find outliers in the numerical variables (you might find outliers in almost all the variables), and check it using visual analysis.

```python
df.info()
```

```python
from pandas.api.types import is_numeric_dtype
numerical_cols = []
for col in df.columns:
    if is_numeric_dtype(df[col].dtype):
        numerical_cols.append(col)
```

```python
numerical_cols = numerical_cols[0:8]
numerical_cols
```

```python
df[numerical_cols].describe(include=["float64"])
```

```python
df.boxplot(figsize = (30,100))
```

```python
Q1 = df[numerical_cols].quantile(0.25)
Q3 = df[numerical_cols].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
print(df[numerical_cols] < (Q1 - 1.5 * IQR)) or (df[numerical_cols] > (Q3 + 1.5 * IQR))
```

#### 2.h. Handle the outliers using the IQR method.

```python
df = df[~((df[numerical_cols] < (Q1 - 1.5 * IQR)) |(df[numerical_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
print(df.shape)
```

```python
df.reset_index(inplace=True)
```

```python
df = df.drop('index', axis=1)
df
```

#### 2.a. Calculate the time taken between od_start_time and od_end_time and keep it as a feature. Drop the original columns, if required.

```python
df['actual_trip_duration'] = pd.to_datetime( df['od_end_time']) - pd.to_datetime( df['od_start_time'])
df['actual_trip_duration'] = df['actual_trip_duration'].dt.seconds/60
df.head()
```

#### 2.b. Compare the difference between Point a. and start_scan_to_end_scan. Do hypothesis testing/ Visual analysis to check.

Hypothesis Testing: start_scan_to_end_scan v/s actual_trip_duration

    ho: start_scan_to_end_scan != actual_trip_duration
    ha: start_scan_to_end_scan == actual_trip_duration

```python
sns.set(rc = {'figure.figsize':(10,10)})
sns.heatmap(df.corr(), cmap="YlGnBu", annot=True)
```

```python
df[['start_scan_to_end_scan', 'actual_trip_duration']].corr()
```

Inference: 'start_scan_to_end_scan' & 'actual_trip_duration'  are highly Correlated.

```python
df["start_scan_to_end_scan"].hist(bins=100)
```

```python
df["actual_trip_duration"].hist(bins=100)
```

```python
sm.qqplot(df['start_scan_to_end_scan'], line = "s")
```

```python
sm.qqplot(df['actual_trip_duration'], line = "s")
```

```python
f, p = stats.ttest_ind(df['start_scan_to_end_scan'], df['actual_trip_duration'], alternative='two-sided')
```
